```python
# ìˆ«ì ë°°ì—´ê³¼ ëª©í‘œê°’ì„ ë§¤ê°œ ë³€ìˆ˜ë¡œ ë°›ëŠ”ë‹¤
def solution(numbers, target): 
    # ë‚´ë¶€dfs : ì´í•¨ìˆ˜ëŠ” idx(í˜„ì œ ì²˜ë¦¬ì¤‘ì¸ ì¸ë±ìŠ¤, current_sumí˜„ì¬ ê°€ì§€ì´ í•©ê³„ë¥¼ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ëŠ”ë‹¤)
    def dfs(idx, current_sum):
        # ì¢…ë£Œ ì¡°ê±´: ëª¨ë“  ìˆ«ìë¥¼ ì²˜ë¦¬í–ˆì„ ë•Œ
        if idx == len(numbers):
            # í˜„ì¬ í•©ê³„ê°€ íƒ€ê²Ÿê³¼ ì¼ì¹˜í•˜ë©´ 1, ì•„ë‹ˆë©´ 0 ë°˜í™˜
            return 1 if current_sum == target else 0
        
        # í˜„ì¬ ìˆ«ìë¥¼ ë”í•˜ëŠ” ê²½ìš°
        plus = dfs(idx + 1, current_sum + numbers[idx]) # ì¬ê·€ í˜¸ì¶œ ìê¸° í•¨ìˆ˜ì—ì„œ ìê¸° í˜¸ì¶œí•˜ëˆˆê±°
        
        # í˜„ì¬ ìˆ«ìë¥¼ ë¹¼ëŠ” ê²½ìš°
        minus = dfs(idx + 1, current_sum - numbers[idx])
        
        # ë‘ ê²½ìš°ì˜ ìˆ˜ë¥¼ í•©ì‚°í•˜ì—¬ ë°˜í™˜
        return plus + minus
    
    # DFS ì‹œì‘ (ì¸ë±ìŠ¤ 0, ì´ˆê¸° í•©ê³„ 0ì—ì„œ ì‹œì‘)
    return dfs(0, 0)
```

- [x] í”„ë¡œê·¸ë˜ë¨¸ìŠ¤ "ì²´ìœ¡ë³µ" (ë ˆë²¨ 1) - ì‰¬ìš´ ê·¸ë¦¬ë”” ì‹¤ì „

```python
# n: ì „ì²´ í•™ìƒìˆ˜ lost: ìƒì–´ë²„ë¦° í•™ìƒ reserve: ì—¬ë²Œì„ ê°€ì§„ í•™ìƒ
def solution(n, lost, reserve):
    # ëª¨ë“  í•™ìƒì´ ì²´ìœ¡ë³µ 1ê°œì”© ê°€ì§€ê³  ìˆë‹¤ê³  ì´ˆê¸°í™”
    student = [1] * n
    
    # ë„ë‚œë‹¹í•œ í•™ìƒì€ ì²´ìœ¡ë³µ -1
    for l in lost:
        student[l-1] -= 1
    
    # ì—¬ë²Œ ê°€ì§„ í•™ìƒì€ ì²´ìœ¡ë³µ +1
    for r in reserve:
        student[r-1] += 1
    
    # ì²´ìœ¡ë³µ ë¹Œë ¤ì£¼ê¸°
    for i in range(n):
        # ì²´ìœ¡ë³µì´ ì—†ëŠ” í•™ìƒì¸ ê²½ìš°
        if student[i] == 0:
            # ì•ë²ˆí˜¸ í•™ìƒì—ê²Œ ë¹Œë¦´ ìˆ˜ ìˆëŠ”ì§€ í™•ì¸ (ì¸ë±ìŠ¤ ë²”ìœ„ ì²´í¬ë„ í•„ìš”)
            if i > 0 and student[i-1] == 2:
                student[i] += 1
                student[i-1] -= 1
            # ë’·ë²ˆí˜¸ í•™ìƒì—ê²Œ ë¹Œë¦´ ìˆ˜ ìˆëŠ”ì§€ í™•ì¸
            elif i < n-1 and student[i+1] == 2:
                student[i] += 1
                student[i+1] -= 1
    
    # ì²´ìœ¡ë³µì´ 1ê°œ ì´ìƒì¸ í•™ìƒ ìˆ˜ ì¹´ìš´íŠ¸
    count = 0
    for s in student:
        if s >= 1:
            count += 1
    
    return count
```

```python
def solution(n):
    answer = 0
    if n % 2 != 0:
        for i in range(1, n+1, 2):
            answer += i
    else:
        for i in range(0, n+1, 2):
            answer += i*i
    return answer
```
