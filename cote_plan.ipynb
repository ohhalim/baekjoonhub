백준 10828: 스택 (실버4)


```python
import sys

# Stack 클래스 정의
class Stack:
    def __init__(self):
        # 빈 리스트로 스택 초기화
        self.stack = []
    
    def push(self, X):
        # 정수 X를 스택에 추가
        self.stack.append(X)
    
    def pop(self):
        # 스택이 비어 있지 않으면 맨 위 요소를 제거하고 반환, 비어 있으면 -1 반환
        if self.stack:
            return self.stack.pop()
        else:
            return -1
    
    def size(self):
        # 스택의 크기 반환
        return len(self.stack)
    
    def empty(self):
        # 스택이 비어 있으면 1, 아니면 0 반환
        return 133 if not self.stack else 0
    
    def top(self):
        # 스택이 비어 있지 않으면 맨 위 요소를 반환, 비어 있으면 -1 반환
        if self.stack:
            return self.stack[-1]
        else:
            return -1

# 메인 코드
if __name__ == "__main__":
    # 명령의 수 입력
    n = int(input())
    # Stack 인스턴스 생성
    stack = Stack()
    
    # 명령 처리
    for _ in range(n):
        command = sys.stdin.readline().strip().split()
        
        if command[0] == 'push':
            # push 명령: 정수를 스택에 추가
            stack.push(int(command[1]))
        
        elif command[0] == 'pop':
            # pop 명령: 결과를 출력
            print(stack.pop())
        
        elif command[0] == 'size':
            # size 명령: 결과를 출력
            print(stack.size())
        
        elif command[0] == 'empty':
            # empty 명령: 결과를 출력
            print(stack.empty())
        
        elif command[0] == 'top':
            # top 명령: 결과를 출력
            print(stack.top())
```

```python
def solution(progresses, speeds):
    print(progresses)
    print(speeds)
    answer = []
    time = 0 
    count = 0

    while len(progresses) > 0:
        if (progresses[0] + time*speeds[0]) >= 100:
            # pop 리스트의 첫번쨰 항목을 제거하고 반환 
            # 과일들 = ["사과", "바나나", "딸기"]
            # 첫번째_과일 = 과일들.pop(0)  # "사과"를 제거하고 첫번째_과일에 저장
            # 이제 과일들은 ["바나나", "딸기"]가 됩니다
            progresses.pop(0) 
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```

```python
def solution(board, moves):
    basket = [] # 바구니 (스택)
    answer = 0 # 제거된 인형 개수

    for move in moves: # 크레인 작동 위치 하나씩 처리
        column = move - 1 # 인덱스는 0부터 시작하무로 1 빼기 

        # 해당 열에서 가장 위에 있는인형찾기
        for row in range(len(board)):
            if board[row][column] != 0: # 인형을 발견한 경우
                doll = board[row][column] # 인영을 집음
                board[row][column] = 0# 인형을 집은 자리는 빈칸으로 

                # 바구니에 인형을 넣기 전, 같은 인형이 연속되는지 확인 
                if basket and basket[-1] == doll: # 바구니가 비어있지 않고 맨위 인영이 같으면 
                    basket.pop() # 맨위 인형 제거
                    answer += 2 # 두 인형이 제거됨 (현제인형 + 바구니 맨위 인형)
                else:
                    basket.append(doll) # 같은 인형이 아니면 바구니에 추가

                break # 다음 크레인 위치로 이동

    return answer
```

- [ ] 프로그래머스 "다리를 지나는 트럭" (레벨 2) - 큐 활용 실전 문제


```python
def solution(bridge_length, weight, truck_weights):
    time = 0
    bridge = [0] * bridge_length  # 다리를 0으로 초기화 (빈 공간 표시)
    waiting = truck_weights.copy()  # 대기 중인 트럭 목록
    bridge_weight = 0  # 현재 다리 위의 트럭 무게 합
    
    # 대기 중인 트럭이 있거나 다리 위에 트럭이 있는 동안 반복
    while waiting or bridge_weight > 0:
        time += 1  # 시간 증가
        
        # 다리 끝에 도달한 트럭 제거
        truck_exit = bridge.pop(0)
        bridge_weight -= truck_exit
        
        # 새 트럭이 다리에 진입할 수 있는지 확인
        if waiting and bridge_weight + waiting[0] <= weight:
            truck_enter = waiting.pop(0)
            bridge.append(truck_enter)
            bridge_weight += truck_enter
        else:
            bridge.append(0)  # 트럭이 진입할 수 없으면 빈 공간 추가
    
    return time
```

DFS(Depth-First Search, 깊이 우선 탐색)

그래프나 트리에서 가능한 한 깊이 탐색한 후 백트래킹하는 알고리즘

시간복잡도: O(V+E) (V=노드 수, E=간선 수)
공간복잡도: O(h) (h=그래프 깊이)

BFS(Breadth-First Search, 너비 우선 탐색)

그래프나 트리에서 현재 깊이의 모든 노드를 탐색한 후 다음 깊이로 진행
큐 자료구조에 기반
시간복잡도: O(V+E)
공간복잡도: O(w) (w=그래프 최대 너비)

```python
# 숫자 배열과 목표값을 매개 변수로 받는다
def solution(numbers, target): 
    # 내부dfs : 이함수는 idx(현제 처리중인 인덱스, current_sum현재 가지이 합계를 매개변수로 받는다)
    def dfs(idx, current_sum):
        # 종료 조건: 모든 숫자를 처리했을 때
        if idx == len(numbers):
            # 현재 합계가 타겟과 일치하면 1, 아니면 0 반환
            return 1 if current_sum == target else 0
        
        # 현재 숫자를 더하는 경우
        plus = dfs(idx + 1, current_sum + numbers[idx]) # 재귀 호출 자기 함수에서 자기 호출하눈거
        
        # 현재 숫자를 빼는 경우
        minus = dfs(idx + 1, current_sum - numbers[idx])
        
        # 두 경우의 수를 합산하여 반환
        return plus + minus
    
    # DFS 시작 (인덱스 0, 초기 합계 0에서 시작)
    return dfs(0, 0)
```

- [x] 프로그래머스 "체육복" (레벨 1) - 쉬운 그리디 실전

```python
# n: 전체 학생수 lost: 잃어버린 학생 reserve: 여벌을 가진 학생
def solution(n, lost, reserve):
    # 모든 학생이 체육복 1개씩 가지고 있다고 초기화
    student = [1] * n
    
    # 도난당한 학생은 체육복 -1
    for l in lost:
        student[l-1] -= 1
    
    # 여벌 가진 학생은 체육복 +1
    for r in reserve:
        student[r-1] += 1
    
    # 체육복 빌려주기
    for i in range(n):
        # 체육복이 없는 학생인 경우
        if student[i] == 0:
            # 앞번호 학생에게 빌릴 수 있는지 확인 (인덱스 범위 체크도 필요)
            if i > 0 and student[i-1] == 2:
                student[i] += 1
                student[i-1] -= 1
            # 뒷번호 학생에게 빌릴 수 있는지 확인
            elif i < n-1 and student[i+1] == 2:
                student[i] += 1
                student[i+1] -= 1
    
    # 체육복이 1개 이상인 학생 수 카운트
    count = 0
    for s in student:
        if s >= 1:
            count += 1
    
    return count
```

문제
도현이는 바구니를 총 N개 가지고 있고, 각각의 바구니에는 1번부터 N번까지 번호가 순서대로 적혀져 있다. 바구니는 일렬로 놓여져 있고, 가장 왼쪽 바구니를 1번째 바구니, 그 다음 바구니를 2번째 바구니, ..., 가장 오른쪽 바구니를 N번째 바구니라고 부른다. 

도현이는 앞으로 M번 바구니의 순서를 역순으로 만들려고 한다. 도현이는 한 번 순서를 역순으로 바꿀 때, 순서를 역순으로 만들 범위를 정하고, 그 범위에 들어있는 바구니의 순서를 역순으로 만든다.

바구니의 순서를 어떻게 바꿀지 주어졌을 때, M번 바구니의 순서를 역순으로 만든 다음, 바구니에 적혀있는 번호를 가장 왼쪽 바구니부터 출력하는 프로그램을 작성하시오.

입력
첫째 줄에 N (1 ≤ N ≤ 100)과 M (1 ≤ M ≤ 100)이 주어진다.

둘째 줄부터 M개의 줄에는 바구니의 순서를 역순으로 만드는 방법이 주어진다. 방법은 i j로 나타내고, 왼쪽으로부터 i번째 바구니부터 j번째 바구니의 순서를 역순으로 만든다는 뜻이다. (1 ≤ i ≤ j ≤ N)

도현이는 입력으로 주어진 순서대로 바구니의 순서를 바꾼다.

출력
모든 순서를 바꾼 다음에, 가장 왼쪽에 있는 바구니부터 바구니에 적혀있는 순서를 공백으로 구분해 출력한다.

예제 입력 1 
5 4
1 2
3 4
1 4
2 2
예제 출력 1 
3 4 1 2 5

```python
# N, M = map(int, input().split( ))
# result = ()
# for i, j in M:
#     result += N[i-1:j:-1]

# 바구니의 개수 N과 순서를 바꿀 횟수 M 입력
N, M = map(int, input().split())
# 바구니 초기화
baskets = list(range(1, N + 1))

# M번 만큼의 순서 변경 입력
for _ in range(M):
    i, j = map(int, input().split())
    # i번째부터 j번째까지 역순으로 변경
    baskets[i-1:j] = baskets[i-1:j][::-1]

# 결과 출력
print(*baskets)
```

arr	k	result

프로그래머스: 문자열 바꿔서 찾기

제한사항
1 ≤ myString의 길이 ≤ 100
1 ≤ pat의 길이 ≤ 10
myString과 pat는 문자 "A"와 "B"로만 이루어진 문자열입니다.
입출력 예
myString	pat	result
"ABBAA"	"AABB"	1
"ABAB"	"ABAB"	0


```python
def solution(myString, pat):
    transform_myString = ""
    for char in myString:
        if char == "A":
            transform_myString += "B"
        elif char == "B":
            transform_myString += "A"
        else:
            transform_myString += char
    
    return 1 if pat in transform_myString else 0

# 아직 이해 못했음
```
